#### 1. Create and plot a random dataset.

```python
# Random data plot

import numpy as np
import matplotlib.pyplot as plt
# define the variables to use

noise = 50
x = np.random.rand(200) * noise
y = np.random.rand(200) * noise

# plot
plt.scatter(x,y, c='r')
plt.show()
```

#### 2. Create and plot a random dataset, your truth.

```python

noise = 50
x = np.random.rand(200) * noise


true_slope , true_intercept = 0, 5

y = true_slope * x + true_intercept

y += np.random.rand(200) * noise

plt.scatter(x , y , c='b')
plt.show()
```

#### 3. Same thing inside a function.

```python
def random_data():
    noise = 50
    x = np.random.rand(200) * noise
    true_slope , true_intercept = 0, 5
    y = true_slope * x + true_intercept
    y += np.random.rand(200) * noise
    return x, y

x_axis, y_axis = random_data()

plt.scatter(x_axis, y_axis, c='g')
plt.show()
```

#### 4. Create 2 datasets inside a class and append them.

```python

class Data:
    def __init__(self,x,y):
        self.noise = 50
        self.x = x
        self.true_slope , self.true_intercept = 0, 5
        self.y = y
        
    def random_data(self):
        self.x = np.random.rand(200) * self.noise
        self.y = self.true_slope * self.x + self.true_intercept
        self.y += np.random.rand(200) * self.noise
        return self.x, self.y 
    
    def append_data(self):
        data1 = self.random_data()
        x_d1 , y_d1 = data1
        data2 = self.random_data()
        x_d2 , y_d2 = data2
        x = np.append(x_d1,x_d2)
        y = np.append(y_d1,y_d2)
        return x, y



appended_result = Data(x,y)

x, y = appended_result.append_data()

plt.scatter(x, y, c='y', label = 'Appended Random Data')
```
